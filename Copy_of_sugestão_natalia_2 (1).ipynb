#**GRUPO_NAO**

```python
GRUPO_NAO_novo['Major Health Topics_2']= GRUPO_NAO_novo['Major Health Topics']
GRUPO_NAO_novo
```

```python
# Nome a ser adicionado
nome_adicional = ', pandemia'

# Função para adicionar o nome adicional
def adicionar_nome(nome):
    return f'{nome} {nome_adicional}'

GRUPO_NAO_novo['Major Health Topics_2'] = GRUPO_NAO_novo['Major Health Topics_2'].apply(adicionar_nome)

GRUPO_NAO_novo
```

```python
import re

for i in range(len(GRUPO_NAO_novo)):

   texto = GRUPO_NAO_novo['Conteudo'][i]

   for j in range(len(GRUPO_NAO_novo['Major Health Topics_2'][i])):

      palavra_alvo=GRUPO_NAO_novo['Major Health Topics_2'][i][j]

      padrao = r'\b(?:\w+\W+){0,15}' + re.escape(palavra_alvo) + r'(?:\W+\w+){0,15}\b'

      correspondencias = re.findall(padrao, texto, re.IGNORECASE)

      if correspondencias != '' :

         GRUPO_NAO_novo['INFORMAÇAO_NAO'][i]=correspondencias
```
